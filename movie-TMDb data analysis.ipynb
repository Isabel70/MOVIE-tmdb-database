```python
df3.columns = [u'Action', u'Adventure', u'Animation', u'Comedy', u'Crime',
       u'Documentary', u'Drama', u'Family', u'Fantasy', u'Foreign', u'History',
       u'Horror', u'Music', u'Mystery', u'Romance', u'Science_Fiction',
       u'TV_Movie', u'Thriller', u'War', u'Western', u'nan', u'revenue_adj']

    
```

```python
total_revenue = []
for i in df3.columns:
    total_revenue.append(df3.query('{}==1'.format(i))['revenue_adj'].sum())
    
```

```python
genre_revenue = pd.DataFrame()
genre_revenue['genre'] = df3.columns
genre_revenue['revenue_sum'] = total_revenue
```

```python
genre_revenue
```

```python
plt.figure(figsize=[20,10])
sns.barplot(data=genre_revenue.sort_values(by='revenue_sum',ascending=False),x='genre',y='revenue_sum')
plt.show()
```
